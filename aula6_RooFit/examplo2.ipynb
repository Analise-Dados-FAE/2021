```c++
// PDF variables
RooRealVar x("x","x",-10.,10.);
RooRealVar y("y","y",0,5);
RooRealVar a("a","a",3.0);
RooRealVar b("b","b",-2.0);
// Generic PDF
RooGenericPdf model("model","Generic PDF",
                    "exp(x*y+a)-b*x",RooArgSet(x,y,a,b));
```

```c++
// Visualizao da PDF
// Plot pdf in observable x
RooPlot* frame = x.frame() ;
model.plotOn(frame) ;
frame->Draw() ;
c1->Draw() ;
```
